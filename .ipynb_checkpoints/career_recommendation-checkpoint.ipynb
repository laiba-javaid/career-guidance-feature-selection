```python
import pandas as pd
import numpy as np
import os
import joblib
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import mutual_info_classif
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
# Load dataset
df = pd.read_csv("data/career_guidance_dataset.csv")
```

```python
# 2. Data Preprocessing
# a) Handle missing values
df = df.dropna()
```

```python
print(df.columns.tolist())
```

```python
# 3. Encode all categorical columns
label_encoders = {}
for column in df.columns:
    if df[column].dtype == 'object':
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        label_encoders[column] = le
```

```python
# 4. Separate features and target
X = df.drop('Recommended Career', axis=1)
y = df['Recommended Career']
features = list(X.columns)
```

```python
#5 Train-test split (70%-30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 6. Feature Selection using Information Gain
info_gain = mutual_info_classif(X_train, y_train)
info_gain_df = pd.DataFrame({'Feature': X.columns, 'Info_Gain': info_gain})
info_gain_df.sort_values(by='Info_Gain', ascending=False, inplace=True)
```

```python
# 5. Evaluation Function
def evaluate_model(X_train_sub, X_test_sub, y_train, y_test):
    results = {}
    for name, model in models.items():
        model.fit(X_train_sub, y_train)
        y_pred = model.predict(X_test_sub)
        results[name] = {
            'accuracy': accuracy_score(y_test, y_pred),
            'precision': precision_score(y_test, y_pred, average='macro', zero_division=0),
            'recall': recall_score(y_test, y_pred, average='macro', zero_division=0)
        }
    return results
```

```python
# 6. Evaluate for multiple thresholds
thresholds = [0.0, 0.01, 0.05, 0.1]
evaluation_results = {}
best_score = 0
best_model = None
best_threshold = None
best_model_name = None
best_selected_features = None

for thresh in thresholds:
    selected_features = info_gain_df[info_gain_df['Info_Gain'] >= thresh]['Feature'].tolist()
    if len(selected_features) == 0:
        continue
    X_train_sub = X_train[selected_features]
    X_test_sub = X_test[selected_features]
    results = evaluate_model(X_train_sub, X_test_sub, y_train, y_test)
    evaluation_results[thresh] = results
    
    for model_name, metrics in results.items():
        if metrics['accuracy'] > best_score:
            best_score = metrics['accuracy']
            best_model = models[model_name]
            best_model_name = model_name
            best_threshold = thresh
            best_selected_features = selected_features
```

```python
# 7. Evaluate baseline (all features, no feature selection)
baseline_results = evaluate_model(X_train, X_test, y_train, y_test)
evaluation_results['All Features'] = baseline_results
```

```python
# 8. Save best model
os.makedirs("model", exist_ok=True)
best_model.fit(X[best_selected_features], y)
joblib.dump(best_model, "model/saved_model.pkl")
joblib.dump(label_encoders, "model/label_encoders.pkl")
joblib.dump(best_selected_features, "model/selected_features.pkl")
```

```python
# 9. Print Results
print("\nEvaluation Results:")
for thresh, results in evaluation_results.items():
    print(f"\nThreshold: {thresh}")
    for model_name, metrics in results.items():
        print(f"  {model_name} - Acc: {metrics['accuracy']:.4f}, Prec: {metrics['precision']:.4f}, Recall: {metrics['recall']:.4f}")

print(f"\n✅ Best Model: {best_model_name} at Threshold: {best_threshold} with Accuracy: {best_score:.4f}")
```

```python
# 10. Visualization
threshold_labels = [str(t) for t in thresholds] + ['All Features']
for metric in ['accuracy', 'precision', 'recall']:
    plt.figure(figsize=(10, 5))
    for model_name in models:
        values = [evaluation_results[t][model_name][metric] if t in evaluation_results else 0 for t in thresholds]
        values.append(baseline_results[model_name][metric])
        plt.plot(threshold_labels, values, label=model_name)
    plt.title(f'{metric.capitalize()} vs. Information Gain Threshold')
    plt.xlabel('Information Gain Threshold')
    plt.ylabel(metric.capitalize())
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f'plots/{metric}_vs_threshold.png')
    plt.show()
```
